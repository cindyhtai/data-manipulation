```python
import pandas as pd
import numpy as np
```

### Problem 1

```python
def get_percent_nans(df, column_name):
    total_rows = df.shape[0]
    total_na = total_rows - df[column_name].count()
    pct = total_na / total_rows * 100
    return pct
```

```python
my_df = pd.read_csv('2019_nCoV_20200121_20200206.csv')
for col in my_df.columns:
    print('The \"{}\" column has {:.2f}% empty values'.format(col, get_percent_nans(my_df, col)))
```

```python
def get_continent(country):
    continents = {'Asia':['Cambodia', 'Hong Kong', 'India', 'Japan', 'Macau', 'Mainland China', 'Malaysia', 'Nepal', 'Philippines', 'Russia', 'Singapore', 'South Korea', 'Sri Lanka', 'Taiwan', 'Thailand', 'United Arab Emirates', 'Vietnam'], 'Europe': ['Belgium', 'Finland', 'France', 'Germany', 'Italy', 'Spain', 'Sweden', 'UK'], 'Australia': ['Australia'], 'North America' : ['Canda', 'United States', 'Mexico'], 'South America' : ['Brazil', 'Colombia'], 'Africa': ['Ivory Coast']}
    for continent in continents:
        if country in continents.get(continent):
            return continent
```

```python
get_continent('UK')
```

### Problem 7

```python
for country in my_df['Country/Region']:
    continent = get_continent(country)
    my_df.loc[(my_df['Country/Region'] == country), 'Continent'] = continent
```
